```python
import pandas as pd
import numpy as np
```

```python
# Save file path to variable
school_csv = "F:camp/hw/hw4/panda-challenge/PyCitySchools/PyCitySchools/Resources/schools_complete.csv"
student_csv = "F:camp/hw/hw4/panda-challenge/PyCitySchools/PyCitySchools/Resources/students_complete.csv"
```

```python
# Read with Pandas
school_df = pd.read_csv(school_csv)
student_df = pd.read_csv(student_csv)
student_df.head()
```

```python
school_df.head()
```

```python
#join the chart
total_df = pd.merge(student_df, school_df, on="school_name",how="outer")
total_df.head()
```

```python
#passing score
passing_score = 70
```

```python
#01 District Summary

# Total Schools
total_school = len(total_df["school_name"].unique())
# Total Students
total_students = len(total_df["Student ID"].unique())
# Total Budget
total_budget = school_df["budget"].sum()
# Average Math Score
avg_math = total_df["math_score"].mean()
# Average Reading Score
avg_reading = total_df["reading_score"].mean()
# % Passing Math (The percentage of students that passed math.)
passing_math = total_df[(total_df["math_score"]>=passing_score)].iloc[:,0].size/total_df.iloc[:,0].size
# % Passing Reading (The percentage of students that passed reading.)
passing_reading = total_df[(total_df["reading_score"]>=passing_score)].iloc[:,0].size/total_df.iloc[:,0].size
# % Overall Passing (The percentage of students that passed math and reading.)
passing_all = total_df[(total_df["math_score"]>=passing_score) & (total_df["reading_score"]>=passing_score)].iloc[:,0].size/total_df.iloc[:,0].size
```

```python
#01 District Summary df
district_s_df = pd.DataFrame(
    {"related_statics":["total_school", "total_students", "total_budget", "avg_math ", "avg_reading", "passing_math ", "passing_reading","passing_all"],
     "data":[total_school,total_students,total_budget,round(avg_math,2) ,round(avg_reading,2), str(round(passing_math*100,2))+"%" ,str(round(passing_reading*100,2))+"%",str(round(passing_all*100,2))+"%"]
     }
)
district_s_df

district_s_df = district_s_df.set_index('related_statics',drop=True)
district_s_df
```

```python
#02 School Summary

# Create a table that highlights the top 5 performing schools based on % Overall Passing. Include:

# School Name
# School Type
# Total Students
# Total School Budget
# Per Student Budget
# Average Math Score
# Average Reading Score
# % Passing Math (The percentage of students that passed math.)
# % Passing Reading (The percentage of students that passed reading.)
# % Overall Passing (The percentage of students that passed math and reading.)
```

```python
#02 Build basic df
school_perf_df = pd.DataFrame(
    {"related_statics":["School Name", 'School Type','total Students', 'Total School Budget','Per Student Budget', 'Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing'],
          }
)

school_perf_df  = school_perf_df.set_index('related_statics',drop=True)
school_perf_df 
```

```python
#02 use for loop to fill in the chart
schools = list(total_df["school_name"].unique())
for target in schools:
    temp_data = total_df.loc[total_df["school_name"] == target, :]
    # basic information
    school_perf_df.loc['School Name',target] = target
    school_perf_df.loc['School Type',target] = temp_data["type"].unique()[0]
    school_perf_df.loc['total Students',target] = temp_data.iloc[:,0].size
    school_perf_df.loc['Total School Budget',target] = int(school_df[school_df["school_name"]==target]["budget"])
    school_perf_df.loc['Per Student Budget',target] =     school_perf_df.loc['Total School Budget',target] /school_perf_df.loc['total Students',target] 
    # performance
    school_perf_df.loc['Average Math Score',target] = temp_data['math_score'].mean()
    school_perf_df.loc['Average Reading Score',target] = temp_data['reading_score'].mean()
    school_perf_df.loc['% Passing Math',target] = round(temp_data[(temp_data["math_score"]>=passing_score)].iloc[:,0].size/temp_data.iloc[:,0].size*100,3)
    school_perf_df.loc['% Passing Reading',target] = round(temp_data[(temp_data["reading_score"]>=passing_score)].iloc[:,0].size/temp_data.iloc[:,0].size*100,3)
    school_perf_df.loc['% Overall Passing',target] = round(temp_data[(temp_data["math_score"]>=passing_score) & (temp_data["reading_score"]>=passing_score)].iloc[:,0].size/temp_data.iloc[:,0].size*100,3)

school_perf_df 
```
